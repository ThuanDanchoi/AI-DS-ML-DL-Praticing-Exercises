```python
summary_data = data['Summary'].value_counts().reset_index()
summary_data.columns = ["Weather Type","Count"]
summary_data
```

```python
missing_data = data.isna().sum()
missing_data
```

```python
total_cells = np.product(data.shape)
total_missing = missing_data.sum()
percent_missing = (total_missing / total_cells) * 100
print(percent_missing)
```

```python
data['Precip Type'] = data['Precip Type'].fillna(data['Precip Type'].value_counts().index[0])
data.isna().sum()
```
