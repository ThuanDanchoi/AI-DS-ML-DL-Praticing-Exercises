```python
data.groupby('Month')['Passenger_Load_Factor'].mean().plot(kind='line')
plt.title('Average Passenger Load Factor Over Time')
plt.ylabel('Passenger Load Factor')
plt.show()
```

```python
data.groupby('City1')['Passenger_Trips'].sum().plot(kind='bar')
plt.title('Total Passenger Trips by City1')
plt.ylabel('Passenger Trips')
plt.show()
```

```python
plt.scatter(data['Distance_GC_(km)'], data['Passenger_Trips'])
plt.title('Distance vs Passenger Trips')
plt.xlabel('Distance (km)')
plt.ylabel('Passenger Trips')
plt.show()

```

```python
data['Passenger_Load_Factor'].plot(kind='hist', bins=20)
plt.title('Distribution of Passenger Load Factor')
plt.xlabel('Load Factor')
plt.show()
```
